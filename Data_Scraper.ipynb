```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import random
import lxml
import numpy
```

```python
url     = "https://free-proxy-list.net/"
test_ip = "http://icanhazip.com"
###  get the HTTP response and construct soup object
r       = requests.get(url)
soup    = BeautifulSoup(r.content, "lxml")
table   = soup.find("table", attrs = {"class": "table table-striped table-bordered"})#.prettify()

proxies   = []
anonymity = []
lim       = 10

### Premier League data lists.
## get_league_stats():
clubs = []
matches = []
wins = []
draws = []
losses = []
goals = []
goals_against = []
form = []
last_5 = []
club_form = []
attend = []
## get_squad_stats():
alphabetical_clubs = []
average_age = []
possession = []
yellow_card = []
red_card = []
## get_goalkeeping():
saves = []
clean_sheets = []
## get_shooting():
shots = []
shots_on_target = []
## get_passing():
attempted_passes = []
complete_passes = []
## get_defensive():
attempted_tackles = []
tackles_won = []
## get_possession():
touches = []
## get_miscellaneous():
offsides = []
fouls = []
interceptions = []
crosses = []
## home/away
clubs_AvsH = []
#######################
home_matches = []
home_wins = []
home_draws = []
home_losses = []
#######################
away_matches = []
away_wins = []
away_draws = []
away_losses = []
```

```python
def get_proxies():
    ''''''' A function that cleans html code from a proxy websites table, to extracts three variables: Ip_Address + Port, and Anonymity (the type of proxy) '''''''
    
    for row in table.find_all("tr")[1:]:
        td = row.find_all("td")
        try:
            ip_address = td[0].text.strip()
            port      = td[1].text.strip()
            host      = f"{ip_address}:{port}"
            proxies.append(host)
            
            anon      = td[4].text
            anonymity.append(anon)

        except IndexError:
            continue
    return proxies, anonymity

def get_league_stats():
    ''''''' A function that '''''''
    
    for x in league_table.tbody.find_all("tr")[0:19]:
        td = x.find_all("td")
        try:
            club = td[0].a.text
            clubs.append(club)
            
            matches_played = td[1].text.strip()
            matches.append(matches_played)
            
            win = td[2].text.strip()
            wins.append(win)
            
            draw = td[3].text.strip()
            draws.append(draw)
            
            loss = td[4].text.strip()
            losses.append(loss)
            
            goal = td[5].text.strip()
            goals.append(goal)
            
            conseed = td[6].text.strip()
            goals_against.append(conseed)
            
            for a in td[13].div("a"):
                club_form.append(a.string)
                
            attendance = td[14].string.strip()
            attend.append(attendance)
        
        except IndexError:
            continue
    return clubs, matches, wins, draws, losses, goals, goals_against, form, attend


def get_squad_stats():
    ''''''' A function that '''''''
    
    for y in Squad_Stats.tbody.find_all("tr")[0:19]:
        td = y.find_all("td")
        try:
            club = td[0].text
            alphabetical_clubs.append(club)
            
            age = td[2].text.strip()
            average_age.append(age)
            
            poss = td[3].text.strip()
            possession.append(poss)
            
            yell = td[13].text.strip()
            yellow_card.append(yell)
            
            red = td[14].text.strip()
            red_card.append(red)
        
        except IndexError:
            continue
    return alphabetical_clubs, age, possession, yellow_card, red_card


def get_goalkeeping():
    ''''''' A function that '''''''
    
    for z in Goalkeeping.tbody.find_all("tr")[0:19]:
        td = z.find_all("td")
        try:
            save = td[9].text.strip()
            saves.append(save)
            
            clean_sheet = td[14].text.strip()
            clean_sheets.append(clean_sheet)

        except IndexError:
            continue
    return saves,  clean_sheets

def get_shooting():
    ''''''' A function that '''''''
    
    for j in Shooting.tbody.find_all("tr")[0:19]:
        td = j.find_all("td")
        try:
            shot = td[4].text.strip()
            shots.append(shot)
            
            sot= td[5].text.strip()
            shots_on_target.append(sot)

        except IndexError:
            continue
    return shots,  shots_on_target

def get_passing():
    ''''''' A function that '''''''
    
    for k in Passing.tbody.find_all("tr")[0:19]:
        td = k.find_all("td")
        try:
            attempt = td[4].text.strip()
            attempted_passes.append(attempt)
            
            complete = td[3].text.strip()
            complete_passes.append(complete)

        except IndexError:
            continue
    return attempted_passes,  complete_passes

def get_defensive():
    ''''''' A function that '''''''
    
    for alpha in Defensive.tbody.find_all("tr")[0:19]:
        td = alpha.find_all("td")
        try:
            tackles = td[4].text.strip()
            attempted_tackles.append(tackles)
            
            tackle_won = td[3].text.strip()
            tackles_won.append(tackle_won)

        except IndexError:
            continue
    return attempted_tackles,  tackles_won

def get_possession():
    ''''''' A function that '''''''
    
    for beta in Possession.tbody.find_all("tr")[0:19]:
        td = beta.find_all("td")
        try:
            touch = td[4].text.strip()
            touches.append(touch)
            
        except IndexError:
            continue
    return touches


def get_miscellaneous():
    ''''''' A function that '''''''
    
    for gamma in Miscellaneous.tbody.find_all("tr")[0:19]:
        td = gamma.find_all("td")
        try:
            foul = td[6].text.strip()
            fouls.append(foul)
            
            off = td[8].text.strip()
            offsides.append(off)
            
            cross = td[9].text.strip()
            crosses.append(cross)
            
            intercept = td[10].text.strip()
            interceptions.append(intercept)
            
        except IndexError:
            continue
    return fouls, offsides, crosses, interceptions

def get_home_away():
    ''''''' A function that '''''''
    
    for delta in home_away.tbody("tr")[0:]:
        td = delta.find_all("td")
        try:
            clubah = td[0].a.text
            clubs_AvsH.append(clubah)
            #################################
            
            home_match = td[1].text.strip()
            home_matches.append(home_match)
            
            home_win = td[2].text.strip()
            home_wins.append(home_win)
            
            home_draw = td[3].text.strip()
            home_draws.append(home_draw)
            
            home_loss = td[4].text.strip()
            home_losses.append(home_loss)
            ###################################
            
            away_match = td[14].text.strip()
            away_matches.append(away_match)
            
            away_win = td[15].text.strip()
            away_wins.append(away_win)
            
            away_draw = td[16].text.strip()
            away_draws.append(away_draw)
            
            away_loss = td[17].text.strip()
            away_losses.append(away_loss)
            
        except IndexError:
            continue
    return 
```
